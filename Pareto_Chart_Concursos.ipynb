```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter


#create DataFrame for reading csv with pandas

df = pd.read_csv('/content/sample_data/ED_VERT_BANCO_DADOS.csv')

#create DataFrame Index

#df.index = df['Índice']

#sort DataFrame by count descending
df = df.sort_values(by='Quantidade', ascending=False)

# Limited index for Pareto Analys
df = df[0:20]

#add column to display cumulative percentage
df['cumperc'] = df['Quantidade'].cumsum()/df['Quantidade'].sum()*100

#define aesthetics for plot
color1 = 'steelblue'
color2 = 'red'
line_size = 4

#create basic bar plot
fig,ax = plt.subplots(figsize = (25,5))
ax.set_title('ED_VERT_DIR_CONST')
ax.bar(df['Índice'], df['Quantidade'], color=color1)

#add cumulative percentage line to plot
ax2 = ax.twinx()
ax2.plot(df['Índice'], df['cumperc'], color=color2, marker="D", ms=line_size)
ax2.yaxis.set_major_formatter(PercentFormatter())

#specify axis colors
ax.tick_params(axis='y', colors=color1)
ax2.tick_params(axis='y', colors=color2)
for tick in ax.get_xticklabels():
    tick.set_rotation(90)

#save image
#plt.savefig('ED_VERT_DIR_CONST.png', format='png', dpi = 600, bbox_inches = 'tight')

#display Pareto chart
plt.show()

```
