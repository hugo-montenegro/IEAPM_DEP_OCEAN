```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from pathlib import Path
```

```python
req = requests.get('https://www.estrategiaconcursos.com.br/blog/notas-corte-controle/')
if req.status_code == 200:
    print('Requisição bem sucedida!')
    content = req.content
```

```python
soup = BeautifulSoup(content, 'html.parser')
table = soup.find(name='table')
```

```python
table_str = str(table)
df = pd.read_html((table_str))[0]
df
```

```python
filepath = Path('C:/Users/hughm/Desktop/out.csv')
filepath.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(filepath)
```
